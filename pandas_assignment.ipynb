# Pandas Assignment
This notebook demonstrates DataFrame creation, filtering, grouping, sorting, and exporting.

```python
import pandas as pd

data = {
    'Name': ['Alice','Bob','Charlie','Diana','Eve'],
    'Age': [24,27,22,32,29],
    'Department': ['HR','Finance','IT','Marketing','HR'],
    'Salary': [45000,54000,50000,62000,47000]
}

df = pd.DataFrame(data)
df
```

## First five rows

```python
df.head()
```

## Summary statistics

```python
df[['Age','Salary']].describe()
```

## Average salary in HR

```python
df[df['Department']=='HR']['Salary'].mean()
```

## Add Bonus column

```python
df['Bonus'] = df['Salary']*0.10
df
```

## Filter age between 25 and 30

```python
df[(df['Age']>=25)&(df['Age']<=30)]
```

## Average salary by department

```python
df.groupby('Department')['Salary'].mean()
```

## Sort by salary and save CSV

```python
sorted_df = df.sort_values(by='Salary')
sorted_df.to_csv('sorted_salary.csv',index=False)
sorted_df
```
