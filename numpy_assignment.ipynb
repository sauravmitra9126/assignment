# NumPy Assignment
This notebook demonstrates array creation, manipulation, statistics, reshaping, and matrix operations using NumPy.

## 1. Create 1D array and compute statistics

```python
import numpy as np

arr = np.arange(1,21)
arr
```

```python
# statistics
sum_val = np.sum(arr)
mean_val = np.mean(arr)
median_val = np.median(arr)
std_val = np.std(arr)

sum_val, mean_val, median_val, std_val
```

```python
# indices greater than 10
np.where(arr > 10)
```

## 2. Create 4x4 2D array

```python
arr2 = np.arange(1,17).reshape(4,4)
arr2
```

```python
# transpose
arr2.T
```

```python
# row and column sums
row_sum = arr2.sum(axis=1)
col_sum = arr2.sum(axis=0)

row_sum, col_sum
```

## 3. Random 3x3 arrays

```python
a = np.random.randint(1,21,(3,3))
b = np.random.randint(1,21,(3,3))

a, b
```

```python
# elementwise operations
a+b, a-b, a*b
```

```python
# dot product
np.dot(a,b)
```

## 4. Reshape and slice

```python
arr3 = np.arange(1,13).reshape(3,4)
arr3
```

```python
# first two rows, last two columns
arr3[:2, -2:]
```
