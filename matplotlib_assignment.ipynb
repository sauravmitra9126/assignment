# Matplotlib Assignment
This notebook demonstrates line plots, bar charts, pie charts, and histograms.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot

```python
x = [1,2,3,4,5]
y = [10,15,25,30,50]

plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

## Bar Graph

```python
students = ['John','Jane','Alice','Bob']
marks = [75,85,60,90]

plt.bar(students,marks,color=['red','blue','green','purple'])
plt.title("Student Marks")
plt.show()
```

## Pie Chart

```python
regions = ['North America','Europe','Asia','Others']
revenue = [45,25,20,10]

explode = [0.1 if r==max(revenue) else 0 for r in revenue]

plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%')
plt.title("Revenue Distribution")
plt.show()
```

## Histogram

```python
data = np.random.randint(1,101,1000)

plt.hist(data, bins=20)
plt.title("Histogram of Random Integers")
plt.show()
```
